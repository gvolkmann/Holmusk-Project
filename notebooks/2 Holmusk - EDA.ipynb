# Data Processing 

```python
df.head()
```

```python
race_dummy = pd.get_dummies(df['race'])
df = pd.concat([df,race_dummy], axis=1)
df.drop(['race'], axis=1, inplace=True)
df.head()
```

```python
resident_status_dummy = pd.get_dummies(df['resident_status'])
df = pd.concat([df,resident_status_dummy], axis=1)
df.drop(['resident_status'], axis=1, inplace=True)
df.head()
```

```python
day_dummy = pd.get_dummies(df['day_of_week'])
df = pd.concat([df,day_dummy], axis=1)
df.drop(['day_of_week'], axis=1, inplace=True)
df.head()
```

```python
month_dummy = pd.get_dummies(df['month'])
df = pd.concat([df,month_dummy], axis=1)
df.drop(['month'], axis=1, inplace=True)
df.head()
```

```python
df['gender'].unique()
```

```python
df["gender"].replace({"Male": 0, "Female": 1}, inplace=True)
df['gender'].unique()
```

```python
df.set_index("patient_id", inplace = True)
df.head()
```

```python
df.drop(columns=['day_of_admission', 'amount_log'], inplace=True)
df.columns
```
